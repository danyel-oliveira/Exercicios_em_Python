```python
#16- Escreva um programa que vá solicitando as idades dos alunos da sala até que todos sejam informados (perguntar ao usuário se deseja
#informar a idade do próximo aluno). Ao final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de 18 anos,
#quantos alunos têm até 18 anos, a média aritmética e a mediana.

idades = []
idade = int(input("Digite a idade do aluno: "))
while idade != 0:
    idades.append(idade) # Indented this line to be part of the while loop
    idade = int(input("Digite a idade do aluno: ")) # Indented this line to be part of the while loop
idades.sort()
n = len(idades)
mais_novo = idades[0]
mais_velho = idades[-1]
maiores_de_18 = len([i for i in idades if i > 18])
menores_de_18 = len([i for i in idades if i <= 18])
media_aritmetica = sum(idades) / n
if n % 2 == 0:
    mediana = (idades[n//2-1] + idades[n//2]) / 2 # Indented this line to be part of the if block
else:
    mediana = idades[n//2] # Indented this line to be part of the else block
print(f"Idade do mais novo: {mais_novo}")
print(f"Idade do mais velho: {mais_velho}")
print(f"Quantidade de alunos com mais de 18 anos: {maiores_de_18}")
print(f"Quantidade de alunos com até 18 anos: {menores_de_18}")
print(f"Média aritmética das idades: {media_aritmetica:.2f}")
print(f"Mediana das idades: {mediana:.2f}")
```
