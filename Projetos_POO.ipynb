```python
# Classe que representa um livro com título, autor e ISBN
class Livro:
    def __init__(self, titulo, autor, isbn):
        self.titulo = titulo          # Título do livro
        self.autor = autor            # Nome do autor
        self.isbn = isbn              # Código ISBN do livro

    def exibir_infor(self):
        # Retorna uma string com as informações do livro
        return f"Título: {self.titulo}, Autor: {self.autor}, ISBN: {self.isbn}"

# Classe que representa um usuário da biblioteca
class Usuario:
    def __init__(self, nome, id_usuario):
        self.nome = nome              # Nome do usuário
        self.id_usuario = id_usuario  # Identificador único do usuário

    def exibir_infor(self):
        # Retorna uma string com as informações do usuário
        return f"Nome: {self.nome}, ID: {self.id_usuario}"

# Classe que representa um empréstimo de livro
class Emprestimo:
    def __init__(self, livro, usuario, data_devolucao):
        self.livro = livro                  # Objeto do tipo Livro
        self.usuario = usuario              # Objeto do tipo Usuario
        self.data_devolucao = data_devolucao  # Data de devolução do livro

    def exibir_infor(self):
        # Retorna uma string com informações sobre o empréstimo
        return f"Livro: {self.livro.titulo}, Usuário: {self.usuario.nome}, Devolução: {self.data_devolucao}"

# Classe principal que gerencia os dados da biblioteca
class Biblioteca:
    def __init__(self):
        self.livros = []       # Lista para armazenar livros
        self.usuarios = []     # Lista para armazenar usuários
        self.emprestimos = []  # Lista para armazenar empréstimos

    def adicionar_livro(self, livro):
        # Adiciona um livro à lista da biblioteca
        self.livros.append(livro)

    def adicionar_usuario(self, usuario):
        # Adiciona um usuário à lista da biblioteca
        self.usuarios.append(usuario)

    def registrar_emprestimo(self, emprestimo):
        # Adiciona um empréstimo à lista de empréstimos
        self.emprestimos.append(emprestimo)

# ---------------- EXEMPLO DE USO ----------------

# Criação de uma instância da biblioteca
biblioteca = Biblioteca()

# Criação de um livro
livro1 = Livro("Dom Quixote", "Miguel de Cervantes", "978-85-336-1573-0")

# Criação de um usuário
usuario1 = Usuario("João Silva", "123")

# Criação de um empréstimo envolvendo o livro e o usuário
emprestimo1 = Emprestimo(livro1, usuario1, "2024-04-30")

# Adicionando os objetos criados à biblioteca
biblioteca.adicionar_livro(livro1)
biblioteca.adicionar_usuario(usuario1)
biblioteca.registrar_emprestimo(emprestimo1)

# Exibindo as informações de livro, usuário e empréstimo
print(biblioteca.livros[0].exibir_infor())
print(biblioteca.usuarios[0].exibir_infor())
print(biblioteca.emprestimos[0].exibir_infor())
```

```python
# Classe que representa um produto da loja
class produto:
    def __init__(self, nome, preco, id_produto):
        self.nome = nome                  # Nome do produto
        self.preco = preco                # Preço do produto
        self.id_produto = id_produto      # ID do produto

    def exibir_infor(self):
        # Retorna as informações formatadas do produto
        return f"Nome: {self.nome}, Preço: R$ {self.preco:.2f}, ID: {self.id_produto}"

# Classe que representa um cliente da loja
class cliente:
    def __init__(self, nome, id_cliente, endereco):  # Corrigido a ordem dos parâmetros
        self.nome = nome                     # Nome do cliente
        self.id_cliente = id_cliente         # ID do cliente
        self.endereço = endereco             # Endereço do cliente

    def exibir_infor(self):
        # Retorna as informações formatadas do cliente
        return f"Nome: {self.nome}, ID: {self.id_cliente}, Endereço: {self.endereço}"

# Classe que representa um pedido feito por um cliente
class pedido:
    def __init__(self, cliente):
        self.cliente = cliente              # Cliente que fez o pedido
        self.produtos = []                  # Lista de produtos adicionados ao pedido
        self.total = 0.0                    # Valor total do pedido

    def adicionar_produto(self, produto):
        # Adiciona um produto ao pedido e atualiza o total
        self.produtos.append(produto)
        self.total += produto.preco

    def exibir_info(self):
        # Exibe as informações do pedido: cliente, produtos e total
        info = f"Cliente: {self.cliente.nome}\nProdutos:\n"
        for produto in self.produtos:
            info += f"- {produto.nome} - R$ {produto.preco:.2f}\n"
        info += f"Total: R$ {self.total:.2f}"
        return info

# Classe principal que gerencia a loja
class Loja:
    def __init__(self):
        self.produtos = []    # Lista de produtos disponíveis na loja
        self.clientes = []    # Lista de clientes cadastrados
        self.pedidos = []     # Lista de pedidos registrados

    def adicionar_produto(self, produto):
        # Adiciona um produto à loja
        self.produtos.append(produto)

    def adicionar_cliente(self, cliente):
        # Adiciona um cliente à loja
        self.clientes.append(cliente)

    def registrar_pedido(self, pedido):
        # Registra um novo pedido
        self.pedidos.append(pedido)

# ---------------- EXEMPLO DE USO ----------------

# Instancia a loja
loja = Loja()

# Cria um produto e um cliente
produto1 = produto("Camiseta", 29.99, "123")
cliente1 = cliente("João Silva", "123456789", "Rua A, 123")

# Cria um pedido e adiciona o produto ao pedido
pedido1 = pedido(cliente1)
pedido1.adicionar_produto(produto1)

# Registra os dados na loja
loja.adicionar_produto(produto1)
loja.adicionar_cliente(cliente1)
loja.registrar_pedido(pedido1)

# Exibe as informações salvas
print(loja.produtos[0].exibir_infor())
print(loja.clientes[0].exibir_infor())
print(loja.pedidos[0].exibir_info())
```

```python
# 1️⃣ Sistema de Gestão de Tarefas 📝

# Classe que representa uma tarefa individual na lista
class Tarefa:
    def __init__(self, nome):
        self.nome = nome        # Nome da tarefa
        self.proxima = None     # Referência para a próxima tarefa (inicialmente None)

# Classe que representa a lista encadeada de tarefas
class ListaEncadeada:
    def __init__(self):
        self.head = None        # Cabeça (início) da lista, começa vazia

    # Adiciona uma nova tarefa no início da lista
    def adicionar_tarefa(self, nome):
        nova_tarefa = Tarefa(nome)        # Cria uma nova tarefa
        nova_tarefa.proxima = self.head   # A nova tarefa aponta para o antigo primeiro item
        self.head = nova_tarefa           # A nova tarefa agora é o primeiro item da lista

    # Remove uma tarefa da lista pelo nome
    def remover_tarefa(self, nome):
        atual = self.head        # Começa no início da lista
        anterior = None          # Variável para armazenar o item anterior
        # Percorre a lista até encontrar a tarefa com o nome correspondente
        while atual and atual.nome != nome:
            anterior = atual
            atual = atual.proxima
        # Se encontrou a tarefa, remove-a da lista
        if atual:
            if anterior:  # Se não for o primeiro item
                anterior.proxima = atual.proxima
            else:         # Se for o primeiro item
                self.head = atual.proxima

    # Mostra todas as tarefas da lista
    def mostrar_tarefas(self):
        atual = self.head
        while atual:
            print(atual.nome)
            atual = atual.proxima

# ---------------- Exemplo de uso ----------------

# Criação da lista de tarefas
lista_tarefa = ListaEncadeada()

# Adiciona algumas tarefas à lista
lista_tarefa.adicionar_tarefa("Estudar Python")
lista_tarefa.adicionar_tarefa("Revisar Código")
lista_tarefa.adicionar_tarefa("Ler sobre POO")

# Mostra todas as tarefas atuais
print("Tarefas iniciais:")
lista_tarefa.mostrar_tarefas()

# Remove uma tarefa específica
lista_tarefa.remover_tarefa("Revisar Código")

# Mostra a lista após a remoção
print("\nTarefas após remover 'Revisar Código':")
lista_tarefa.mostrar_tarefas()
```

```python
# Importando o módulo abc para usar classes abstratas
from abc import ABC, abstractmethod

# Classe abstrata base para dispositivos eletrônicos
class DispostivoEletronico(ABC):
    @abstractmethod
    def ligar(self):
        pass

    @abstractmethod
    def desligar(self):
        pass

    @abstractmethod
    def exibir_info(self):
        pass

# Classe concreta que representa uma Televisão, herdando da classe abstrata
class Televisao(DispostivoEletronico):
    def __init__(self):
        self.ligada = False  # Estado inicial desligado
        self.volume = 15     # Volume padrão

    def ligar(self):
        self.ligada = True
        print("Televisão ligada.")

    def desligar(self):
        self.ligada = False
        print("Televisão desligada.")

    def exibir_info(self):
        # Exibe o status da TV (ligada ou desligada) e o volume atual
        print(f"Status da TV: {'Ligada' if self.ligada else 'Desligada'}, Volume: {self.volume}")

    def aumentar_volume(self):
        # Só aumenta o volume se a TV estiver ligada
        if self.ligada:
            self.volume += 1
            print(f"Volume aumentado para {self.volume}.")
        else:
            print("Ligue a televisão para ajustar o volume.")

    def diminuir_volume(self):
        # Diminui o volume se a TV estiver ligada e o volume for maior que 0
        if self.ligada and self.volume > 0:
            self.volume -= 1
            print(f"Volume diminuído para {self.volume}.")
        else:
            print("Volume mínimo atingido.")

# Classe concreta que representa um Smartphone, também herda de DispositivoEletronico
class Smartphone(DispostivoEletronico):
    def __init__(self):
        self.ligado = False
        self.aplicativo_aberto = None  # Nenhum app aberto inicialmente

    def ligar(self):
        self.ligado = True
        print("Smartphone ligado.")

    def desligar(self):
        self.ligado = False
        print("Smartphone desligado.")

    def exibir_info(self):
        # Exibe o status (ligado ou desligado) e o aplicativo aberto, se houver
        status = "Ligado" if self.ligado else "Desligado"
        app_info = f", Aplicativo Aberto: {self.aplicativo_aberto}" if self.aplicativo_aberto else ""
        print(f"Status do Smartphone: {status}{app_info}.")

    def abrir_aplicativo(self, nome_app):
        # Abre o aplicativo apenas se o smartphone estiver ligado
        if self.ligado:
            self.aplicativo_aberto = nome_app
            print(f"Abrindo o aplicativo: {nome_app}")
        else:
            print("Ligue o smartphone para abrir aplicativos.")

    def exibir_status(self):
        # Método auxiliar que reutiliza exibir_info()
        self.exibir_info()

    def fechar_aplicativo(self):
        # Fecha o app se o smartphone estiver ligado e com app aberto
        if self.ligado and self.aplicativo_aberto:
            print(f"Fechado o aplicativo: {self.aplicativo_aberto}")
            self.aplicativo_aberto = None
        elif self.ligado:
            print("Nenhum aplicativo aberto.")
        else:
            print("Ligue o smartphone.")

# ------------------- TESTE DOS DISPOSITIVOS -------------------

tv = Televisao()            # Instancia uma TV
smartphone = Smartphone()  # Instancia um Smartphone

# Teste das funcionalidades da TV
tv.ligar()
tv.aumentar_volume()
tv.exibir_info()
tv.desligar()

print("-" * 20)  # Separador visual

# Teste das funcionalidades do Smartphone
smartphone.ligar()
smartphone.abrir_aplicativo("WhatsApp")
smartphone.exibir_status()  # Exibe status com app aberto
smartphone.fechar_aplicativo()
smartphone.desligar()
```

```python
# Jogo de Aventura baseado em Texto 🎮

# Classe base que representa um personagem genérico
class Personagem:
    def __init__(self, nome, saude):
        self.nome = nome              # Nome do personagem
        self.saude = saude            # Nível de saúde do personagem

    def apresentar(self):
        # Método para exibir uma apresentação do personagem
        print(f"Eu sou {self.nome}, tenho {self.saude} de saúde.")

# Classe que representa um guerreiro, herda de Personagem
class Guerreiro(Personagem):
    def __init__(self, nome):
        # O guerreiro sempre inicia com 100 de saúde
        super().__init__(nome, saude=100)

    def atacar(self):
        # Ação exclusiva do guerreiro
        print(f"{self.nome} ataca com sua espada!")

    def defender(self):
        # Ação de defesa do guerreiro
        print(f"{self.nome} se prepara para defender o ataque!")

# Classe que representa um mago, herda de Personagem
class Mago(Personagem):
    def __init__(self, nome):
        # O mago inicia com 200 de saúde
        super().__init__(nome, saude=200)

    def lançar_feitiço(self):
        # Ação exclusiva do mago
        print(f"{self.nome} lança um feitiço poderoso!")

    def defender(self):
        # Ação de defesa do mago
        print(f"{self.nome} conjura um escudo mágico para se defender!")

# Classe que representa uma cena do jogo
class Cena:
    def __init__(self, descricao):
        self.descricao = descricao    # Texto que descreve a cena

    def mostrar(self):
        # Exibe a descrição da cena
        print(self.descricao)

# Classe principal do jogo que controla o fluxo
class Jogo:
    def __init__(self):
        self.personagem = None        # Personagem escolhido pelo jogador
        self.cenas = []               # Lista de cenas do jogo

    def criar_personagem(self):
        # Solicita ao jogador que escolha e crie um personagem
        tipo = input("Escolha seu personagem (guerreiro/mago): ").strip().lower()
        nome = input("Digite o nome do seu personagem: ")
        if tipo == "guerreiro":
            self.personagem = Guerreiro(nome)
        elif tipo == "mago":
            self.personagem = Mago(nome)
        else:
            # Caso o tipo seja inválido, repete a escolha
            print("Tipo de personagem inválido.")
            self.criar_personagem()

    def adicionar_cena(self, cena):
        # Adiciona uma nova cena ao jogo
        self.cenas.append(cena)

    def jogar(self):
        # Inicia o jogo, cria o personagem e percorre as cenas
        self.criar_personagem()
        for cena in self.cenas:
            cena.mostrar()
            # Jogador escolhe uma ação
            acao = input("O que você deseja fazer? (atacar/lançar feitiço/defender): ").strip().lower()
            if acao == "atacar" and isinstance(self.personagem, Guerreiro):
                self.personagem.atacar()
            elif acao == "lançar feitiço" and isinstance(self.personagem, Mago):
                self.personagem.lançar_feitiço()
            elif acao == "defender":
                self.personagem.defender()
            else:
                print("Ação inválida ou personagem não pode realizar essa ação.")

# Criando o jogo
jogo = Jogo()

# Adicionando algumas cenas com descrições
jogo.adicionar_cena(Cena("Você está em uma floresta sombria."))
jogo.adicionar_cena(Cena("Você encontra um dragão feroz!"))

# Iniciando o jogo
jogo.jogar()
```

```python
 #1
class A:
    # Método construtor que inicializa o atributo x
    def __init__(self, x):
        self.x = x  # Atributo x é atribuído ao valor passado como argumento

    # Método para exibir o valor de x
    def exibir(self):
        print(f"Valor de x em A: {self.x}")  # Imprime o valor de x

# Definição da classe B que herda de A
class B(A):
    # Método construtor que inicializa os atributos x e y
    def __init__(self, x, y):
        super().__init__(x)  # Chama o construtor da classe A para inicializar x
        self.y = y  # Atributo y é atribuído ao valor passado como argumento

    # Método para exibir os valores de x e y
    def exibir(self):
        super().exibir()  # Chama o método exibir da classe A para mostrar x
        print(f"Valor de y em B: {self.y}")  # Imprime o valor de y

# Criação de uma instância da classe B com x = 10 e y = 20
obj_b = B(10, 20)

# Chama o método exibir da instância obj_b para mostrar os valores de x e y
obj_b.exibir()  # Saída esperada: "Valor de x em A: 10" e "Valor de y em B: 20"
```

```python
# 2
class Calculadora:
    # Método para somar dois números
    def somar(self, a, b):
        return a + b  # Retorna a soma de a e b

# Cria uma instância da classe Calculadora
calc = Calculadora()

# Chama o método somar com os valores 5 e 3 e armazena o resultado
resultado = calc.somar(5, 3)

# Imprime o resultado da soma
print(resultado)  # Saída esperada: 8
```

```python
#3 Definindo a classe Animal
class Animal:
    # Método que será chamado para fazer um som genérico
    def fazer_som(self):
        print("Som genérico")  # Imprime "Som genérico"

# Definindo a classe Cachorro, que herda da classe Animal
class Cachorro(Animal):
    # Sobrescreve o método fazer_som da classe Animal para fazer o som específico do cachorro
    def fazer_som(self):
        print("Au au!")  # Imprime "Au au!", som típico de cachorro

# Criando um objeto da classe Animal
animal1 = Animal()  # Este objeto usará o método fazer_som da classe Animal

# Criando um objeto da classe Cachorro
animal2 = Cachorro()  # Este objeto usará o método fazer_som sobrescrito da classe Cachorro

# Chamando o método fazer_som no objeto animal1 (que é da classe Animal)
animal1.fazer_som()  # Exibe "Som genérico", pois animal1 é da classe Animal

# Chamando o método fazer_som no objeto animal2 (que é da classe Cachorro)
animal2.fazer_som()  # Exibe "Au au!", pois animal2 é da classe Cachorro e o método foi sobrescrito
```

```python
#4 Definindo a classe Contador
class Contador:
    # Atributo de classe contador, que será compartilhado por todas as instâncias
    contador = 0

    # Método construtor (inicializa o objeto)
    def __init__(self):
        Contador.contador += 1  # Sempre que um novo objeto é criado, incrementa o contador da classe

    # Método para exibir o valor atual do contador
    def exibir_contador(self):
        print(f"Contador: {Contador.contador}")  # Exibe o valor atual do contador da classe

# Criando o primeiro objeto da classe Contador
obj1 = Contador()  # Isso chama o __init__ e incrementa Contador.contador de 0 para 1

# Criando o segundo objeto da classe Contador
obj2 = Contador()  # Isso chama o __init__ e incrementa Contador.contador de 1 para 2

# Chamando o método exibir_contador do objeto obj1
obj1.exibir_contador()  # Exibe "Contador: 2", pois o contador foi incrementado duas vezes (uma para cada objeto)
```

```python
#5 Definindo a classe Pessoa
class Pessoa:
    # Método construtor (inicializa o objeto) com um parâmetro 'nome'
    def __init__(self, nome):
        self._nome = nome  # Atribui o valor do parâmetro 'nome' ao atributo protegido '_nome'

    # Método para acessar o valor do atributo protegido '_nome'
    def get_nome(self):
        return self._nome  # Retorna o valor do atributo '_nome'

# Criando um objeto da classe Pessoa com o nome "Carlos"
pessoa = Pessoa("Carlos")

# Chamando o método get_nome para obter e imprimir o nome da pessoa
print(pessoa.get_nome())  # Exibe o nome "Carlos"
```
