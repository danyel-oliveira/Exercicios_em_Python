```python
# Classe que representa um livro com t√≠tulo, autor e ISBN
class Livro:
    def __init__(self, titulo, autor, isbn):
        self.titulo = titulo          # T√≠tulo do livro
        self.autor = autor            # Nome do autor
        self.isbn = isbn              # C√≥digo ISBN do livro

    def exibir_infor(self):
        # Retorna uma string com as informa√ß√µes do livro
        return f"T√≠tulo: {self.titulo}, Autor: {self.autor}, ISBN: {self.isbn}"

# Classe que representa um usu√°rio da biblioteca
class Usuario:
    def __init__(self, nome, id_usuario):
        self.nome = nome              # Nome do usu√°rio
        self.id_usuario = id_usuario  # Identificador √∫nico do usu√°rio

    def exibir_infor(self):
        # Retorna uma string com as informa√ß√µes do usu√°rio
        return f"Nome: {self.nome}, ID: {self.id_usuario}"

# Classe que representa um empr√©stimo de livro
class Emprestimo:
    def __init__(self, livro, usuario, data_devolucao):
        self.livro = livro                  # Objeto do tipo Livro
        self.usuario = usuario              # Objeto do tipo Usuario
        self.data_devolucao = data_devolucao  # Data de devolu√ß√£o do livro

    def exibir_infor(self):
        # Retorna uma string com informa√ß√µes sobre o empr√©stimo
        return f"Livro: {self.livro.titulo}, Usu√°rio: {self.usuario.nome}, Devolu√ß√£o: {self.data_devolucao}"

# Classe principal que gerencia os dados da biblioteca
class Biblioteca:
    def __init__(self):
        self.livros = []       # Lista para armazenar livros
        self.usuarios = []     # Lista para armazenar usu√°rios
        self.emprestimos = []  # Lista para armazenar empr√©stimos

    def adicionar_livro(self, livro):
        # Adiciona um livro √† lista da biblioteca
        self.livros.append(livro)

    def adicionar_usuario(self, usuario):
        # Adiciona um usu√°rio √† lista da biblioteca
        self.usuarios.append(usuario)

    def registrar_emprestimo(self, emprestimo):
        # Adiciona um empr√©stimo √† lista de empr√©stimos
        self.emprestimos.append(emprestimo)

# ---------------- EXEMPLO DE USO ----------------

# Cria√ß√£o de uma inst√¢ncia da biblioteca
biblioteca = Biblioteca()

# Cria√ß√£o de um livro
livro1 = Livro("Dom Quixote", "Miguel de Cervantes", "978-85-336-1573-0")

# Cria√ß√£o de um usu√°rio
usuario1 = Usuario("Jo√£o Silva", "123")

# Cria√ß√£o de um empr√©stimo envolvendo o livro e o usu√°rio
emprestimo1 = Emprestimo(livro1, usuario1, "2024-04-30")

# Adicionando os objetos criados √† biblioteca
biblioteca.adicionar_livro(livro1)
biblioteca.adicionar_usuario(usuario1)
biblioteca.registrar_emprestimo(emprestimo1)

# Exibindo as informa√ß√µes de livro, usu√°rio e empr√©stimo
print(biblioteca.livros[0].exibir_infor())
print(biblioteca.usuarios[0].exibir_infor())
print(biblioteca.emprestimos[0].exibir_infor())
```

```python
# Classe que representa um produto da loja
class produto:
    def __init__(self, nome, preco, id_produto):
        self.nome = nome                  # Nome do produto
        self.preco = preco                # Pre√ßo do produto
        self.id_produto = id_produto      # ID do produto

    def exibir_infor(self):
        # Retorna as informa√ß√µes formatadas do produto
        return f"Nome: {self.nome}, Pre√ßo: R$ {self.preco:.2f}, ID: {self.id_produto}"

# Classe que representa um cliente da loja
class cliente:
    def __init__(self, nome, id_cliente, endereco):  # Corrigido a ordem dos par√¢metros
        self.nome = nome                     # Nome do cliente
        self.id_cliente = id_cliente         # ID do cliente
        self.endere√ßo = endereco             # Endere√ßo do cliente

    def exibir_infor(self):
        # Retorna as informa√ß√µes formatadas do cliente
        return f"Nome: {self.nome}, ID: {self.id_cliente}, Endere√ßo: {self.endere√ßo}"

# Classe que representa um pedido feito por um cliente
class pedido:
    def __init__(self, cliente):
        self.cliente = cliente              # Cliente que fez o pedido
        self.produtos = []                  # Lista de produtos adicionados ao pedido
        self.total = 0.0                    # Valor total do pedido

    def adicionar_produto(self, produto):
        # Adiciona um produto ao pedido e atualiza o total
        self.produtos.append(produto)
        self.total += produto.preco

    def exibir_info(self):
        # Exibe as informa√ß√µes do pedido: cliente, produtos e total
        info = f"Cliente: {self.cliente.nome}\nProdutos:\n"
        for produto in self.produtos:
            info += f"- {produto.nome} - R$ {produto.preco:.2f}\n"
        info += f"Total: R$ {self.total:.2f}"
        return info

# Classe principal que gerencia a loja
class Loja:
    def __init__(self):
        self.produtos = []    # Lista de produtos dispon√≠veis na loja
        self.clientes = []    # Lista de clientes cadastrados
        self.pedidos = []     # Lista de pedidos registrados

    def adicionar_produto(self, produto):
        # Adiciona um produto √† loja
        self.produtos.append(produto)

    def adicionar_cliente(self, cliente):
        # Adiciona um cliente √† loja
        self.clientes.append(cliente)

    def registrar_pedido(self, pedido):
        # Registra um novo pedido
        self.pedidos.append(pedido)

# ---------------- EXEMPLO DE USO ----------------

# Instancia a loja
loja = Loja()

# Cria um produto e um cliente
produto1 = produto("Camiseta", 29.99, "123")
cliente1 = cliente("Jo√£o Silva", "123456789", "Rua A, 123")

# Cria um pedido e adiciona o produto ao pedido
pedido1 = pedido(cliente1)
pedido1.adicionar_produto(produto1)

# Registra os dados na loja
loja.adicionar_produto(produto1)
loja.adicionar_cliente(cliente1)
loja.registrar_pedido(pedido1)

# Exibe as informa√ß√µes salvas
print(loja.produtos[0].exibir_infor())
print(loja.clientes[0].exibir_infor())
print(loja.pedidos[0].exibir_info())
```

```python
# 1Ô∏è‚É£ Sistema de Gest√£o de Tarefas üìù

# Classe que representa uma tarefa individual na lista
class Tarefa:
    def __init__(self, nome):
        self.nome = nome        # Nome da tarefa
        self.proxima = None     # Refer√™ncia para a pr√≥xima tarefa (inicialmente None)

# Classe que representa a lista encadeada de tarefas
class ListaEncadeada:
    def __init__(self):
        self.head = None        # Cabe√ßa (in√≠cio) da lista, come√ßa vazia

    # Adiciona uma nova tarefa no in√≠cio da lista
    def adicionar_tarefa(self, nome):
        nova_tarefa = Tarefa(nome)        # Cria uma nova tarefa
        nova_tarefa.proxima = self.head   # A nova tarefa aponta para o antigo primeiro item
        self.head = nova_tarefa           # A nova tarefa agora √© o primeiro item da lista

    # Remove uma tarefa da lista pelo nome
    def remover_tarefa(self, nome):
        atual = self.head        # Come√ßa no in√≠cio da lista
        anterior = None          # Vari√°vel para armazenar o item anterior
        # Percorre a lista at√© encontrar a tarefa com o nome correspondente
        while atual and atual.nome != nome:
            anterior = atual
            atual = atual.proxima
        # Se encontrou a tarefa, remove-a da lista
        if atual:
            if anterior:  # Se n√£o for o primeiro item
                anterior.proxima = atual.proxima
            else:         # Se for o primeiro item
                self.head = atual.proxima

    # Mostra todas as tarefas da lista
    def mostrar_tarefas(self):
        atual = self.head
        while atual:
            print(atual.nome)
            atual = atual.proxima

# ---------------- Exemplo de uso ----------------

# Cria√ß√£o da lista de tarefas
lista_tarefa = ListaEncadeada()

# Adiciona algumas tarefas √† lista
lista_tarefa.adicionar_tarefa("Estudar Python")
lista_tarefa.adicionar_tarefa("Revisar C√≥digo")
lista_tarefa.adicionar_tarefa("Ler sobre POO")

# Mostra todas as tarefas atuais
print("Tarefas iniciais:")
lista_tarefa.mostrar_tarefas()

# Remove uma tarefa espec√≠fica
lista_tarefa.remover_tarefa("Revisar C√≥digo")

# Mostra a lista ap√≥s a remo√ß√£o
print("\nTarefas ap√≥s remover 'Revisar C√≥digo':")
lista_tarefa.mostrar_tarefas()
```

```python
# Importando o m√≥dulo abc para usar classes abstratas
from abc import ABC, abstractmethod

# Classe abstrata base para dispositivos eletr√¥nicos
class DispostivoEletronico(ABC):
    @abstractmethod
    def ligar(self):
        pass

    @abstractmethod
    def desligar(self):
        pass

    @abstractmethod
    def exibir_info(self):
        pass

# Classe concreta que representa uma Televis√£o, herdando da classe abstrata
class Televisao(DispostivoEletronico):
    def __init__(self):
        self.ligada = False  # Estado inicial desligado
        self.volume = 15     # Volume padr√£o

    def ligar(self):
        self.ligada = True
        print("Televis√£o ligada.")

    def desligar(self):
        self.ligada = False
        print("Televis√£o desligada.")

    def exibir_info(self):
        # Exibe o status da TV (ligada ou desligada) e o volume atual
        print(f"Status da TV: {'Ligada' if self.ligada else 'Desligada'}, Volume: {self.volume}")

    def aumentar_volume(self):
        # S√≥ aumenta o volume se a TV estiver ligada
        if self.ligada:
            self.volume += 1
            print(f"Volume aumentado para {self.volume}.")
        else:
            print("Ligue a televis√£o para ajustar o volume.")

    def diminuir_volume(self):
        # Diminui o volume se a TV estiver ligada e o volume for maior que 0
        if self.ligada and self.volume > 0:
            self.volume -= 1
            print(f"Volume diminu√≠do para {self.volume}.")
        else:
            print("Volume m√≠nimo atingido.")

# Classe concreta que representa um Smartphone, tamb√©m herda de DispositivoEletronico
class Smartphone(DispostivoEletronico):
    def __init__(self):
        self.ligado = False
        self.aplicativo_aberto = None  # Nenhum app aberto inicialmente

    def ligar(self):
        self.ligado = True
        print("Smartphone ligado.")

    def desligar(self):
        self.ligado = False
        print("Smartphone desligado.")

    def exibir_info(self):
        # Exibe o status (ligado ou desligado) e o aplicativo aberto, se houver
        status = "Ligado" if self.ligado else "Desligado"
        app_info = f", Aplicativo Aberto: {self.aplicativo_aberto}" if self.aplicativo_aberto else ""
        print(f"Status do Smartphone: {status}{app_info}.")

    def abrir_aplicativo(self, nome_app):
        # Abre o aplicativo apenas se o smartphone estiver ligado
        if self.ligado:
            self.aplicativo_aberto = nome_app
            print(f"Abrindo o aplicativo: {nome_app}")
        else:
            print("Ligue o smartphone para abrir aplicativos.")

    def exibir_status(self):
        # M√©todo auxiliar que reutiliza exibir_info()
        self.exibir_info()

    def fechar_aplicativo(self):
        # Fecha o app se o smartphone estiver ligado e com app aberto
        if self.ligado and self.aplicativo_aberto:
            print(f"Fechado o aplicativo: {self.aplicativo_aberto}")
            self.aplicativo_aberto = None
        elif self.ligado:
            print("Nenhum aplicativo aberto.")
        else:
            print("Ligue o smartphone.")

# ------------------- TESTE DOS DISPOSITIVOS -------------------

tv = Televisao()            # Instancia uma TV
smartphone = Smartphone()  # Instancia um Smartphone

# Teste das funcionalidades da TV
tv.ligar()
tv.aumentar_volume()
tv.exibir_info()
tv.desligar()

print("-" * 20)  # Separador visual

# Teste das funcionalidades do Smartphone
smartphone.ligar()
smartphone.abrir_aplicativo("WhatsApp")
smartphone.exibir_status()  # Exibe status com app aberto
smartphone.fechar_aplicativo()
smartphone.desligar()
```

```python
# Jogo de Aventura baseado em Texto üéÆ

# Classe base que representa um personagem gen√©rico
class Personagem:
    def __init__(self, nome, saude):
        self.nome = nome              # Nome do personagem
        self.saude = saude            # N√≠vel de sa√∫de do personagem

    def apresentar(self):
        # M√©todo para exibir uma apresenta√ß√£o do personagem
        print(f"Eu sou {self.nome}, tenho {self.saude} de sa√∫de.")

# Classe que representa um guerreiro, herda de Personagem
class Guerreiro(Personagem):
    def __init__(self, nome):
        # O guerreiro sempre inicia com 100 de sa√∫de
        super().__init__(nome, saude=100)

    def atacar(self):
        # A√ß√£o exclusiva do guerreiro
        print(f"{self.nome} ataca com sua espada!")

    def defender(self):
        # A√ß√£o de defesa do guerreiro
        print(f"{self.nome} se prepara para defender o ataque!")

# Classe que representa um mago, herda de Personagem
class Mago(Personagem):
    def __init__(self, nome):
        # O mago inicia com 200 de sa√∫de
        super().__init__(nome, saude=200)

    def lan√ßar_feiti√ßo(self):
        # A√ß√£o exclusiva do mago
        print(f"{self.nome} lan√ßa um feiti√ßo poderoso!")

    def defender(self):
        # A√ß√£o de defesa do mago
        print(f"{self.nome} conjura um escudo m√°gico para se defender!")

# Classe que representa uma cena do jogo
class Cena:
    def __init__(self, descricao):
        self.descricao = descricao    # Texto que descreve a cena

    def mostrar(self):
        # Exibe a descri√ß√£o da cena
        print(self.descricao)

# Classe principal do jogo que controla o fluxo
class Jogo:
    def __init__(self):
        self.personagem = None        # Personagem escolhido pelo jogador
        self.cenas = []               # Lista de cenas do jogo

    def criar_personagem(self):
        # Solicita ao jogador que escolha e crie um personagem
        tipo = input("Escolha seu personagem (guerreiro/mago): ").strip().lower()
        nome = input("Digite o nome do seu personagem: ")
        if tipo == "guerreiro":
            self.personagem = Guerreiro(nome)
        elif tipo == "mago":
            self.personagem = Mago(nome)
        else:
            # Caso o tipo seja inv√°lido, repete a escolha
            print("Tipo de personagem inv√°lido.")
            self.criar_personagem()

    def adicionar_cena(self, cena):
        # Adiciona uma nova cena ao jogo
        self.cenas.append(cena)

    def jogar(self):
        # Inicia o jogo, cria o personagem e percorre as cenas
        self.criar_personagem()
        for cena in self.cenas:
            cena.mostrar()
            # Jogador escolhe uma a√ß√£o
            acao = input("O que voc√™ deseja fazer? (atacar/lan√ßar feiti√ßo/defender): ").strip().lower()
            if acao == "atacar" and isinstance(self.personagem, Guerreiro):
                self.personagem.atacar()
            elif acao == "lan√ßar feiti√ßo" and isinstance(self.personagem, Mago):
                self.personagem.lan√ßar_feiti√ßo()
            elif acao == "defender":
                self.personagem.defender()
            else:
                print("A√ß√£o inv√°lida ou personagem n√£o pode realizar essa a√ß√£o.")

# Criando o jogo
jogo = Jogo()

# Adicionando algumas cenas com descri√ß√µes
jogo.adicionar_cena(Cena("Voc√™ est√° em uma floresta sombria."))
jogo.adicionar_cena(Cena("Voc√™ encontra um drag√£o feroz!"))

# Iniciando o jogo
jogo.jogar()
```

```python
 #1
class A:
    # M√©todo construtor que inicializa o atributo x
    def __init__(self, x):
        self.x = x  # Atributo x √© atribu√≠do ao valor passado como argumento

    # M√©todo para exibir o valor de x
    def exibir(self):
        print(f"Valor de x em A: {self.x}")  # Imprime o valor de x

# Defini√ß√£o da classe B que herda de A
class B(A):
    # M√©todo construtor que inicializa os atributos x e y
    def __init__(self, x, y):
        super().__init__(x)  # Chama o construtor da classe A para inicializar x
        self.y = y  # Atributo y √© atribu√≠do ao valor passado como argumento

    # M√©todo para exibir os valores de x e y
    def exibir(self):
        super().exibir()  # Chama o m√©todo exibir da classe A para mostrar x
        print(f"Valor de y em B: {self.y}")  # Imprime o valor de y

# Cria√ß√£o de uma inst√¢ncia da classe B com x = 10 e y = 20
obj_b = B(10, 20)

# Chama o m√©todo exibir da inst√¢ncia obj_b para mostrar os valores de x e y
obj_b.exibir()  # Sa√≠da esperada: "Valor de x em A: 10" e "Valor de y em B: 20"
```

```python
# 2
class Calculadora:
    # M√©todo para somar dois n√∫meros
    def somar(self, a, b):
        return a + b  # Retorna a soma de a e b

# Cria uma inst√¢ncia da classe Calculadora
calc = Calculadora()

# Chama o m√©todo somar com os valores 5 e 3 e armazena o resultado
resultado = calc.somar(5, 3)

# Imprime o resultado da soma
print(resultado)  # Sa√≠da esperada: 8
```

```python
#3 Definindo a classe Animal
class Animal:
    # M√©todo que ser√° chamado para fazer um som gen√©rico
    def fazer_som(self):
        print("Som gen√©rico")  # Imprime "Som gen√©rico"

# Definindo a classe Cachorro, que herda da classe Animal
class Cachorro(Animal):
    # Sobrescreve o m√©todo fazer_som da classe Animal para fazer o som espec√≠fico do cachorro
    def fazer_som(self):
        print("Au au!")  # Imprime "Au au!", som t√≠pico de cachorro

# Criando um objeto da classe Animal
animal1 = Animal()  # Este objeto usar√° o m√©todo fazer_som da classe Animal

# Criando um objeto da classe Cachorro
animal2 = Cachorro()  # Este objeto usar√° o m√©todo fazer_som sobrescrito da classe Cachorro

# Chamando o m√©todo fazer_som no objeto animal1 (que √© da classe Animal)
animal1.fazer_som()  # Exibe "Som gen√©rico", pois animal1 √© da classe Animal

# Chamando o m√©todo fazer_som no objeto animal2 (que √© da classe Cachorro)
animal2.fazer_som()  # Exibe "Au au!", pois animal2 √© da classe Cachorro e o m√©todo foi sobrescrito
```

```python
#4 Definindo a classe Contador
class Contador:
    # Atributo de classe contador, que ser√° compartilhado por todas as inst√¢ncias
    contador = 0

    # M√©todo construtor (inicializa o objeto)
    def __init__(self):
        Contador.contador += 1  # Sempre que um novo objeto √© criado, incrementa o contador da classe

    # M√©todo para exibir o valor atual do contador
    def exibir_contador(self):
        print(f"Contador: {Contador.contador}")  # Exibe o valor atual do contador da classe

# Criando o primeiro objeto da classe Contador
obj1 = Contador()  # Isso chama o __init__ e incrementa Contador.contador de 0 para 1

# Criando o segundo objeto da classe Contador
obj2 = Contador()  # Isso chama o __init__ e incrementa Contador.contador de 1 para 2

# Chamando o m√©todo exibir_contador do objeto obj1
obj1.exibir_contador()  # Exibe "Contador: 2", pois o contador foi incrementado duas vezes (uma para cada objeto)
```

```python
#5 Definindo a classe Pessoa
class Pessoa:
    # M√©todo construtor (inicializa o objeto) com um par√¢metro 'nome'
    def __init__(self, nome):
        self._nome = nome  # Atribui o valor do par√¢metro 'nome' ao atributo protegido '_nome'

    # M√©todo para acessar o valor do atributo protegido '_nome'
    def get_nome(self):
        return self._nome  # Retorna o valor do atributo '_nome'

# Criando um objeto da classe Pessoa com o nome "Carlos"
pessoa = Pessoa("Carlos")

# Chamando o m√©todo get_nome para obter e imprimir o nome da pessoa
print(pessoa.get_nome())  # Exibe o nome "Carlos"
```
